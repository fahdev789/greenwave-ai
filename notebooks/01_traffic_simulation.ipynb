### Importing libraries

```python
# üìÅ Imports
import matplotlib.pyplot as plt
from environment.simple_simulator import TrafficSimulator
from utils.metrics import compute_total_wait_time
```

#### Executing the Simulator to analyze the queue of Traffic

```python
# üîß Initialize Simulation
sim = TrafficSimulator(sim_time=100)  # Simulate for 100 seconds

# üìä Track queues
queue_history = {
    'N': [],
    'S': [],
    'E': [],
    'W': []
}

# ‚ñ∂Ô∏è Run Simulation
for _ in range(sim.sim_time):
    sim.step()
    for d in queue_history:
        queue_history[d].append(len(sim.queues[d]))
```

#### Visualizing the queue lengths in all the four lanes

```python
# üìà Plot Queue Lengths
plt.figure(figsize=(12, 6))
for direction in queue_history:
    plt.plot(queue_history[direction], label=f"{direction} queue")

plt.title("Queue Lengths of No. of Cars Over Time")
plt.xlabel("Waiting Time (seconds)")
plt.ylabel("Queue Length of Cars")
plt.legend()
plt.grid(True)
plt.show()

# ‚úÖ Final Evaluation
total_wait = compute_total_wait_time(sim.queues)
print(f"Total waiting time at the end: {total_wait}")
```
